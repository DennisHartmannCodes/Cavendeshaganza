```python
# importing module
from pandas import *
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
 
# reading CSV file
data = read_csv("End-Deflection run 1 whacky.csv")
 
# converting column data to list
times = data['t(s)'].tolist()
Position = data['Position(cm)'].tolist()
Error = data['Error in Position (cm)'].tolist()
 
# printing list data
plt.scatter(times,Position, color='b')
plt.errorbar(times, Position, yerr=Error, ecolor='r', fmt="o", barsabove="True")
plt.title('End-Deflection Method')
plt.xlabel('Time (s)')
plt.ylabel('Position (cm)')
plt.savefig('EndDeflectrun1.png')
```

```python
# reading CSV file
data = read_csv("Cavendeshaganza - Accelration run 1.csv")
 
# converting column data to list
timesquared = data['t^2(s^2)'].tolist()
Position = data['Position (cm)+1.25'].tolist()
Error = data['Error in Position(cm)'].tolist()
 
# printing list data
plt.scatter(timesquared,Position, color='b')
plt.errorbar(timesquared, Position, yerr=Error, ecolor='r', fmt="o", barsabove="True")
plt.xticks(np.arange(0, 75, step=5))  # Set label locations.
plt.xlim(39,61)
plt.title('Acceleration Method')
plt.xlabel('Time^2 (s^2)')
plt.ylabel('Position (cm)')
plt.savefig('Accelrun1.png')
```

```python
# reading CSV file
data = read_csv("Cavendeshaganza - End-Deflection Run 2.csv")
 
# converting column data to list
times = data['t(s)'].tolist()
times=times[:83]
Position = data['Position(cm)'].tolist()
Position=Position[:83]
Error = data['Error in Position (cm)'].tolist()
Error=Error[:83]
 
# printing list data
plt.scatter(times,Position, color='b')
plt.errorbar(times, Position, yerr=Error, ecolor='r', fmt="o", barsabove="True")
plt.title('End-Deflection Method')
plt.xlabel('Time (s)')
plt.ylabel('Position (cm)')

max_index=Position.index(max(Position[5:]))
print(max_index)
min_index=Position.index(min(Position[5:]))
print(min_index)

slope, intercept, r_value, p_value, std_err = stats.linregress(times,Position)

print(intercept)
b=[intercept] *83
plt.plot(times,b)

plt.savefig('EndDeflectrun2.png')

```
