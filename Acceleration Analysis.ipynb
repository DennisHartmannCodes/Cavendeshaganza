```python
# importing module
from pandas import *
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import re
```

```python
# reading CSV file
data = read_csv("Cavendeshaganza - Accelration run 1.csv")
 
# converting column data to list
timesquared = data['t^2(s^2)'].tolist()
timesquared=timesquared[6:18]
Position = data['Position (cm)+1.25'].tolist()
Position=Position[6:18]
Error = data['Error in Position(cm)'].tolist()
Error=Error[6:18]

for i in range(len(timesquared)):
    timesquared[i] = re.sub(",", "", timesquared[i])
    timesquared[i]=int(timesquared[i])

# printing list data
plt.scatter(timesquared,Position, color='b')
plt.errorbar(timesquared, Position, yerr=Error, ecolor='r', fmt="o", barsabove="True")
plt.title('Acceleration Method')
plt.xlabel('Time^2 (s^2)')
plt.ylabel('Position (cm)')
#plt.xticks(np.arange(0, 61, step=10))  # Set label locations.
#plt.xlim(2,19)

# for i in range(len(timesquared)):
#     timesquared[i] = re.sub(",", "", timesquared[i])
#     timesquared[i]=int(timesquared[i])

slope1, intercept, r_value, p_value, std_err = stats.linregress(timesquared,Position)
timesquaredarr=np.asarray(timesquared)
print(timesquared)
print(std_err)
plt.plot(timesquared, (timesquaredarr*slope1)+intercept)
plt.savefig('Vid1trim.png')
print(slope1, '+/-', std_err)
```
